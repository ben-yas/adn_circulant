# Train model
## with metadata

```python
def split_data(X, y, metadata_to_keep):

    X_train = X[X['train_test'] == 'train'].reset_index(drop=True)
    X_train = pd.merge(X_train, metadata[['Code'] + metadata_to_keep], how='left', left_on='patient_id', right_on='Code')
    not_nan_idx = X_train[~X_train.isnull().any(axis=1)].index.tolist()
    X_train = X_train.loc[not_nan_idx, :]
    
    y_train = y.loc[y['train_test'] == 'train', 'relapse'].reset_index(drop=True)
    y_train = y_train[not_nan_idx]
    X_train = X_train[[c for c in X_train.columns if c not in ['Code', 'patient_id', 'train_test', 'relapse']]]

    X_test = X[X['train_test'] == 'test'].reset_index(drop=True)
    X_test = pd.merge(X_test, metadata[['Code'] + metadata_to_keep], how='left', left_on='patient_id', right_on='Code')
    not_nan_idx = X_test[~X_test.isnull().any(axis=1)].index.tolist()
    X_test = X_test.loc[not_nan_idx, :]
    
    y_test = y.loc[y['train_test'] == 'test', 'relapse'].reset_index(drop=True)
    y_test = y_test[not_nan_idx]
    X_test = X_test[[c for c in X_test.columns if c not in ['Code', 'patient_id', 'train_test', 'relapse']]]
    
    return X_train, y_train, X_test, y_test
```

```python
from sklearn.metrics import accuracy_score, precision_recall_fscore_support
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler

def train(X_train, y_train, X_test, y_test):
    ss = StandardScaler().fit(X_train)
    X_train_scaled = ss.transform(X_train)
    X_test_scaled = ss.transform(X_test)
    
    
    lr = LogisticRegression(C=0.01, class_weight='balanced', random_state=5)

    lr.fit(X_train_scaled, y_train)
    y_train_pred = lr.predict(X_train_scaled)

    train_acc = accuracy_score(y_train_pred, y_train)
    train_scores = precision_recall_fscore_support(y_train_pred, y_train, labels=[0, 1])

    y_test_pred = lr.predict(X_test_scaled)
    test_acc = accuracy_score(y_test_pred, y_test)
    test_scores = precision_recall_fscore_support(y_test_pred, y_test, labels=[0, 1])
    
    print(f'train accuracy : {round(100 * train_acc, 1)}')
    print(f'test accuracy : {round(100 * test_acc, 1)}')
    return train_acc, test_acc
```

```python
from itertools import combinations
metadata_columns = ['sex', 'age', 'pub', 'htum', 'meta_bis', 'cpa_diagnostic']

results = []

for r in range(0, len(metadata_columns) + 1):
    for cols in combinations(metadata_columns, r):
        print(f'******************** {cols}')
        X_train, y_train, X_test, y_test = split_data(X, y, list(cols))
        print(X_train.shape)
        print(X_test.shape)
        train_acc, test_acc = train(X_train, y_train, X_test, y_test)
        results.append((cols, train_acc, test_acc))
```
